```python
import mod1
print(mod1.add(11,22))
print(mod1.sub(33,22))
```

```python
with open("mod2.py","w") as f:
    data="""def mul(a,b):
    return a*b
def div(a,b):
    return a/b"""
    f.write(data)
```

```python
import mod2
print(mod2.mul(3,4))
print(mod2.div(8,4))
```

```python
import mod1 , mod2
print(mod1.add(11,22))
print(mod1.sub(33,22))
print(mod2.mul(3,4))
print(mod2.div(8,4))
```

```python
# import mod1 as m1
# import mod2 as m2
import mod1 as m1 , mod2 as m2
print(m1.add(11,22))
print(m1.sub(33,22))
print(m2.mul(3,4))
print(m2.div(8,4))
```

```python
from mod1 import add , sub
print(add(33,44))
```

```python
from mod2 import *
print(mul(4,5))
```
