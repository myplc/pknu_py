## 클래스나 변수등을 포함한 모듈

```python
with open("mod3.py","w") as f:
    text = """PI = 3.141592\n\nclass Math:\n\tdef solv(self,r):\n\t\treturn PI * (r**2)\n\ndef add(a,b):\n\treturn a+b"""     
    f.write(text)
```

```python
import mod3
# 변수
print(mod3.PI)
# 클래스
a = mod3.Math()
print(a.solv(10))
# 함수
print(mod3.add(mod3.PI, 4.4))
```
