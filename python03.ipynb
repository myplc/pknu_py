```python
me = 'Python'
compA = ['JAVA','CSS','Javascript','C','MySQL']
compB = ['JAVA','Python','Javascript','C','MySQL']
compC = ['JAVA','Python','HTML','MySQL']
if me in compA:
    print("A회사 합격입니다.")
else:
    print("A회사 불합격입니다.")
if me in compB:
    print("B회사 합격입니다.")
else:
    print("B회사 불합격입니다.")
if me in compC:
    print("C회사 합격입니다.")
else:
    print("C회사  불합격입니다.")
```

```python
# 방법1 if만 사용
점수 = int(input("점수를 입력하세요.") or 77)
grade = 'F'
if 90 <= 점수: grade = 'A'
if 80 <= 점수 < 90: grade = 'B'
if 70 <= 점수 < 80: grade = 'C'
if 60 <= 점수 < 70: grade = 'D'    
print(f"{grade} 학점입니다.")
```

```python
# 방법2 if, elif, else 사용
score = int(input("점수를 입력하세요.") or 37)
if 90 <= score : grade = 'A'
elif 80 <= score: grade = 'B'
elif 70 <= score: grade = 'C'
elif 60 <= score: grade = 'D'
else: grade = 'F'
print(f"{grade} 학점입니다.")
```

```python
# 사용자로부터 온도 입력받기
temperature = float(input("현재 온도를 입력하세요 (°C): "))

# 조건문으로 계절 구분
if 15 <= temperature < 25:
    print("지금은 봄입니다 🌸")
elif temperature >= 25:
    print("지금은 여름입니다 🌞")
elif 10 <= temperature < 15:
    print("지금은 가을입니다 🍂")
else:
    print("지금은 겨울입니다 ❄️")
```

## while 문

```python
num = 10
while num:
    num-=1
    print(num, "나화났어!")    
```

```python
treeHit = 0
while treeHit < 10:
    treeHit = treeHit +1
    print("나무를 %d번 찍었습니다." % treeHit)
    if treeHit == 10:
        print("나무 넘어갑니다.")
```

```python
namu = 10
while namu:    
    namu -=1
    print(f"나무를 {10 - namu}번 찍었습니다.")
print("나무 넘어갑니다.")
```

```python
from IPython.display import clear_output
prompt = """
선택해 주세요.
    1. Add
    2. Del
    3. List
    4. Quit
"""
number = 0
while number != 4:
    print(prompt)
    number = int(input())
    clear_output(wait=True) # 이전 질문 지우기
print("시스템이 종료되었습니다.")
```
