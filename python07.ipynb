# 내장함수

```python
# abs *
print(abs(-3))
print(abs(-1.2))
```

```python
# all *
print(all([1,2,3]))
print(all([11,22,0,44,55]))
print(all([])) # 특이사항
```

```python
# any *
print(any([1,2,3,0]))
print(any([0,"",[],False]))
print(any([]))
```

```python
# 아스키코드
print(chr(65))
print(chr(44032))
# 유니코드 - UTF -8(인코딩방식)
print(ord('A'))
print(ord('가'))
```
