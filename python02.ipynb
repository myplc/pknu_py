### 리스트 연산,수정,삭제하기

```python
a = [11,22,33]
b = [44,55,66]
# 리스트 더하기
print(a+b)
# 리스트 반복하기
c = a * 3
print(c)
len(c)
```

```python
# 리스트 값 수정하기
a = [11,22,33]
a[2] = 44
print(a)
# a[3] = 55 # 없는 요소를 수정하는 것은 에러
# print(a)
# a[1] = '' # 빈칸 수정 삽입
# a[1] = 0  # 0으로 수정 삽입
# 리스트 값 삭제하기
del a[1]
print(a)
```

```python
# 리스트값 슬라이싱으로 삭제
print(c)
del c[:4]
print(c)
```

```python
# sorted 응용
fruit = ['banana','kiwi','orange','melon','pear','apple','strawberry','blueberry']
result = sorted(fruit,key=len)
result2 = sorted(fruit,key=len,reverse=True)
print(fruit)
print(result)
print(result2)
```

```python
print('과일을 먹어볼까?')
for r in result:
    if r == 'banana':
        print(r + ' 못먹겠어! 웩!')
    if r != 'banana':
        print(r + ' 냠냠 맛있다.')
print('와 ~ 다먹었다')
```

```python
# 장바구니
bag = []
```

```python
# Create 생성
inp = input('필요한 항목을 넣어주세요.')
bag.append(inp)
print(f'장바구니에는 {bag}들이 있습니다.')
```

```python
# FIFO 출력(큐 출력)
hand = bag.pop(0)
print(f'{hand}를 꺼내서 장바구니에 {bag}이(가) 남았습니다.')
```

```python
# 선택 출력
bag = ['호박', '감자', '소주', '맥주' , '피자']
print(f"바구니에 {bag} 물품이 있습니다.")
bring = input('꺼내고 싶은 물품을 입력하세요.')
hand2 = bag.pop(bag.index(bring))
print(f"바구니에서 {hand2} 꺼내어 {bag}이 남았습니다.")
```
